```python
import polars as pl 
from dbconfig import engine
print('Environment ready!')
```

```python
df = pl.read_csv("inflation_rates.csv")
```

```python
df = df.select(
        pl.col('Observation').alias('year'),
        pl.col('OBS_STATUS').alias('inflation_rate')
        )
```

```python
df.head()
```

```python
df = df.with_columns(
    pl.col("inflation_rate").round(2)
)
```

```python
df.head()
```

```python
df = df.drop_nulls()
```

```python
df
```

```python
df.describe()
```

```python
df = df.with_columns(
    pl.col("year").cast(pl.Int32)
)
```

```python
df.describe()
```

```python
df.schema
```

```python
df.write_database(
    table_name="analysis.inflation",
    connection=engine,
    if_table_exists="replace",
)
```

```python
pl.read_database(
        query = """
        select table_name,
        column_name
        from information_schema.columns
        where table_schema = 'analysis'
        order by table_name, ordinal_position;
        """, connection = engine
        )
```
